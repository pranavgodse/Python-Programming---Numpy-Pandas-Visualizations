Inner Join: Common data points

Left Join: Common data points and also all the data points in the left table only

Right Join: Common data points and also all the data points in the right table only

Outer Join: Fetches the data points available in both the tables

```python
# create a dataframe
df_sales = pd.DataFrame({ 'ID':[101,102,103,104,105,106],
                          'Brand': ['Apple', 'Canon', 'Samsung', 'Nikon', 'Sony','MI'],
                          'Quantity_sold':[234,344,345,262,356,450] })

# print the dataframe
df_sales
```

```python
pd.merge(df_product,df_sales,on =['Brand','ID'],how='left')
```

```python
pd.merge(df_product,df_sales,on =['Brand','ID'],how='right')
```

```python
pd.merge(df_product,df_sales,on =['Brand','ID'],how='inner')
```

```python
pd.merge(df_product,df_sales,on =['Brand','ID'],how='outer')
```
